```python
# Count occurrences of each PK_TEUNA_FIKT
counts = dfk['pk_teuna_fikt'].value_counts()

# Identify which values are unique and which are repeated
unique_ids = counts[counts == 1].index
shared_ids = counts[counts > 1].index

# Split the DataFrame
dfself = dfk[dfk['pk_teuna_fikt'].isin(unique_ids)]
dfs = dfk[dfk['pk_teuna_fikt'].isin(shared_ids)]

# Optional: print how many rows in each
print(f"Rows with unique PK_TEUNA_FIKT (dfself): {len(dfself)}")
print(f"Rows with shared PK_TEUNA_FIKT (dfs): {len(dfs)}")
```

```python

# Count values in 'MOTORCYCLE' column
count_self = dfself['MOTORCYCLE'].value_counts()
count_s = dfs['MOTORCYCLE'].value_counts()

# Print counts
print("Counts in dfself:")
print(count_self)
print("\nCounts in dfs:")
print(count_s)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# dfself plot
axes[0].bar(count_self.index.astype(str), count_self.values, color='skyblue')
axes[0].set_title("dfself: MOTORCYCLE Counts")
axes[0].set_xlabel("MOTORCYCLE")
axes[0].set_ylabel("Count")

# dfs plot
axes[1].bar(count_s.index.astype(str), count_s.values, color='salmon')
axes[1].set_title("dfs: MOTORCYCLE Counts")
axes[1].set_xlabel("MOTORCYCLE")

plt.tight_layout()
plt.show()
```
