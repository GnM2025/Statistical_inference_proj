```python
import pandas as pd
from scipy.stats import chi2_contingency, norm
import numpy as np

# --- Chi-Squared Test for Motorcycle Crash Proportions (Fatal vs. Non-Fatal) ---
print("--- Chi-Squared Test for Motorcycle Crash Proportions ---")

# Define the motorcycle crash data for Israel and New York
# Non-fatal crashes are calculated by subtracting fatal crashes from the total.
israel_moto_fatal = 78
israel_moto_total = 7146
israel_moto_non_fatal = israel_moto_total - israel_moto_fatal

ny_moto_fatal = 76
ny_moto_total = 5965
ny_moto_non_fatal = ny_moto_total - ny_moto_fatal

# Create a contingency table for motorcycle crashes
contingency_table_moto = [
    [israel_moto_fatal, israel_moto_non_fatal],
    [ny_moto_fatal, ny_moto_non_fatal]
]

# Print the contingency table for clarity
print("Motorcycle Crash Contingency Table:")
print(f"               Fatal Crashes    Non-Fatal Crashes")
print(f"Israel          {israel_moto_fatal}              {israel_moto_non_fatal}")
print(f"New York        {ny_moto_fatal}              {ny_moto_non_fatal}")

# Perform the Chi-squared test
chi2_moto, p_value_moto, degrees_of_freedom_moto, expected_frequencies_moto = chi2_contingency(contingency_table_moto)

# Print the results
print("\n--- Chi-Squared Test Results ---")
print(f"Chi-squared Statistic: {chi2_moto:.4f}")
print(f"P-value: {p_value_moto:.10f}")
print(f"Degrees of Freedom: {degrees_of_freedom_moto}")
print(f"Expected Frequencies Table:\n{np.round(expected_frequencies_moto, 2)}")

# Interpret the results
alpha = 0.05
print("\n--- Interpretation ---")
print(f"Significance Level (alpha): {alpha}")
if p_value_moto < alpha:
    print(f"The p-value ({p_value_moto:.10f}) is less than {alpha}.")
    print("Conclusion: The difference in motorcycle fatal crash rates is statistically significant.")
else:
    print(f"The p-value ({p_value_moto:.10f}) is greater than or equal to {alpha}.")
    print("Conclusion: The difference in motorcycle fatal crash rates is not statistically significant.")


# --- Z-Test for Rates (Fatal Crashes per 100,000 Motorcycles) ---
print("\n\n--- Z-Test for Fatal Crash Rates per 100,000 Motorcycles ---")

# Define new data points for motorcycle population
israel_motorcycles = 160000
ny_motorcycles = 350000

# Calculate fatal crash rates per 100,000 motorcycles
israel_moto_rate_per_100k = (israel_moto_fatal / israel_motorcycles) * 100000
ny_moto_rate_per_100k = (ny_moto_fatal / ny_motorcycles) * 100000

print(f"Israel's Motorcycle Fatal Crash Rate: {israel_moto_rate_per_100k:.2f} per 100,000 motorcycles")
print(f"New York's Motorcycle Fatal Crash Rate: {ny_moto_rate_per_100k:.2f} per 100,000 motorcycles")

# Calculate the pooled rate
pooled_rate_moto = (israel_moto_fatal + ny_moto_fatal) / (israel_motorcycles + ny_motorcycles)

# Calculate the standard error of the difference
se_moto = np.sqrt(pooled_rate_moto * ((1 / israel_motorcycles) + (1 / ny_motorcycles)))

# Calculate the Z-statistic
z_statistic_moto = (israel_moto_rate_per_100k / 100000 - ny_moto_rate_per_100k / 100000) / se_moto

# Calculate the p-value from the Z-statistic (two-tailed test)
p_value_z_test_moto = 2 * (1 - norm.cdf(abs(z_statistic_moto)))

# Print the Z-test results
print("\n--- Z-Test Results ---")
print(f"Z-statistic: {z_statistic_moto:.4f}")
print(f"P-value: {p_value_z_test_moto:.10f}")

# Interpret the Z-test results
print("\n--- Interpretation ---")
print(f"Significance Level (alpha): {alpha}")
if p_value_z_test_moto < alpha:
    print(f"The p-value ({p_value_z_test_moto:.10f}) is less than {alpha}.")
    print("Conclusion: The difference in motorcycle fatal crash rates per 100,000 motorcycles is statistically significant.")
else:
    print(f"The p-value ({p_value_z_test_moto:.10f}) is greater than or equal to {alpha}.")
    print("Conclusion: The difference in motorcycle fatal crash rates per 100,000 motorcycles is not statistically significant.")


# --- Motorcycle to Death Ratio ---
print("\n\n--- Fatal Motorcycle Crashes per Motorcycle ---")

# Calculate the fatal crash to motorcycle ratio for Israel
israel_ratio = israel_moto_fatal / israel_motorcycles
print(f"Israel's ratio: {israel_ratio:.6f} (fatal crashes per motorcycle)")

# Calculate the fatal crash to motorcycle ratio for New York
ny_ratio = ny_moto_fatal / ny_motorcycles
print(f"New York's ratio: {ny_ratio:.6f} (fatal crashes per motorcycle)")

# Calculate the relative difference
relative_difference = israel_ratio / ny_ratio
print(f"\nA fatal motorcycle crash was {relative_difference:.2f} times more likely in Israel than in New York, based on these ratios.")
6
```

```python
# --- Two-sample Poisson rate ratio test + plots ---

import numpy as np
import matplotlib.pyplot as plt

# Inputs: person-level serious+fatal counts from steps above
x_IL = int(710)
x_NY = int(NY_count_SF)

# Exposures (registered motorcycles) ‚Äî override with your best 2023 official counts if available
E_IL = 160_000   # Israel, default from your scripts
E_NY = 350_000   # NYC, default from your scripts

# Poisson rate ratio (RR) and z-test
rate_IL = x_IL / E_IL
rate_NY = x_NY / E_NY
RR = rate_IL / rate_NY if rate_NY > 0 else np.nan
# Standard error of log(RR) for Poisson counts with known exposures:
# Var(log(rate)) ~ 1/x, so Var(log(RR)) ~ 1/x_IL + 1/x_NY
se_logRR = np.sqrt((1/x_IL) + (1/x_NY)) if (x_IL>0 and x_NY>0) else np.nan
z = np.log(RR) / se_logRR if (se_logRR and np.isfinite(se_logRR)) else np.nan
from scipy.stats import norm
p_two_sided = 2 * (1 - norm.cdf(abs(z))) if np.isfinite(z) else np.nan

# 95% CI for RR
alpha = 0.05
zcrit = norm.ppf(1 - alpha/2)
ci_lo = np.exp(np.log(RR) - zcrit*se_logRR) if np.isfinite(se_logRR) else np.nan
ci_hi = np.exp(np.log(RR) + zcrit*se_logRR) if np.isfinite(se_logRR) else np.nan

print("\n=== Poisson Rate Comparison: Serious+Fatal motorcycle casualties (persons) ===")
print(f"Exposure (IL, NY): {E_IL:,}, {E_NY:,}")
print(f"Counts   (IL, NY): {x_IL:,}, {x_NY:,}")
print(f"Rates per 100,000 (IL, NY): {rate_IL*1e5:,.2f}, {rate_NY*1e5:,.2f}")
print(f"Rate Ratio (IL/NY): {RR:.3f}  (95% CI: {ci_lo:.3f}‚Äì{ci_hi:.3f})")
print(f"z = {z:.3f},  p-value = {p_two_sided:.4g}")

# --- Plots ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# A) Counts
axes[0].bar(['Israel', 'New York'], [x_IL, x_NY], color=['#1f77b4', '#ff7f0e'])
axes[0].set_title('Serious+Fatal Motorcycle Casualties (count), 2023')
axes[0].set_ylabel('Number of persons')

# B) Rates with 95% CI per 100,000
rates = np.array([rate_IL*1e5, rate_NY*1e5])
# CI for rate_i ‚âà using Poisson exact or normal approx. Use normal approx on log(rate) => CI on rate:
def rate_ci(count, exposure, zcrit=1.96):
    if count == 0:
        # 97.5% upper for 0 is ~3.69; use exact-ish bound; lower is 0
        return 0.0, (3.69/exposure)*1e5
    se_log = 1/np.sqrt(count)
    lo = (count/exposure)*np.exp(-zcrit*se_log)
    hi = (count/exposure)*np.exp(+zcrit*se_log)
    return lo*1e5, hi*1e5
ci_bars = np.array([rate_ci(x_IL, E_IL), rate_ci(x_NY, E_NY)])
yerr = np.abs(ci_bars.T - rates)

axes[1].bar(['Israel', 'New York'], rates, yerr=yerr, capsize=6, color=['#1f77b4', '#ff7f0e'])
axes[1].set_title('Serious+Fatal Motorcycle Casualty Rate (per 100k), 2023')
axes[1].set_ylabel('Rate per 100,000 motorcycles')
plt.tight_layout()
plt.show()

# --- Text conclusion helper ---
if np.isfinite(p_two_sided):
    if p_two_sided < 0.05:
        print(f"\nConclusion: The rate of serious+fatal motorcycle casualties is statistically different between IL and NY (p={p_two_sided:.4g}).")
        if RR > 1:
            print(f"Specifically, IL is higher by a factor of {RR:.2f} ({ci_lo:.2f}‚Äì{ci_hi:.2f} 95% CI).")
        else:
            print(f"Specifically, IL is lower by a factor of {1/RR:.2f} ({1/ci_hi:.2f}‚Äì{1/ci_lo:.2f} inverted 95% CI).")
    else:
        print(f"\nConclusion: No statistically significant difference in rates (p={p_two_sided:.4g}).")
else:
    print("\nConclusion: Insufficient information to compute the test (zero counts or missing severity in NY people table).")
```

```python
# =======================
# NYC 2023: Self vs Multi, using full injury/death columns
# =======================

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, fisher_exact, mannwhitneyu
from statsmodels.stats.contingency_tables import Table2x2
from statsmodels.stats.proportion import proportion_confint
import statsmodels.api as sm

# ---------- 0) CRASHES: clean & restrict to 2023 ----------
crashes = df1.copy()
crashes.columns = crashes.columns.str.strip()
crashes['CRASH DATE'] = pd.to_datetime(crashes['CRASH DATE'], errors='coerce')
crashes_2023 = crashes.loc[crashes['CRASH DATE'].dt.year == 2023].copy()

# Ensure all injury/death columns exist & numeric
col_inj_kill_all = [
    'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
    'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
    'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
    'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
]
for c in col_inj_kill_all:
    if c not in crashes_2023.columns:
        crashes_2023[c] = 0
    crashes_2023[c] = pd.to_numeric(crashes_2023[c], errors='coerce').fillna(0)

# Prefer totals from sub-components (to avoid double counting)
crashes_2023['inj_sum_sub'] = (
    crashes_2023['NUMBER OF PEDESTRIANS INJURED'] +
    crashes_2023['NUMBER OF CYCLIST INJURED'] +
    crashes_2023['NUMBER OF MOTORIST INJURED']
)
crashes_2023['kill_sum_sub'] = (
    crashes_2023['NUMBER OF PEDESTRIANS KILLED'] +
    crashes_2023['NUMBER OF CYCLIST KILLED'] +
    crashes_2023['NUMBER OF MOTORIST KILLED']
)

# If sub-components exist, use them; otherwise fall back to persons totals
crashes_2023['total_injured'] = np.where(
    crashes_2023[['NUMBER OF PEDESTRIANS INJURED','NUMBER OF CYCLIST INJURED','NUMBER OF MOTORIST INJURED']].sum(axis=1).notna(),
    crashes_2023['inj_sum_sub'],
    crashes_2023['NUMBER OF PERSONS INJURED']
)
crashes_2023['total_killed'] = np.where(
    crashes_2023[['NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST KILLED']].sum(axis=1).notna(),
    crashes_2023['kill_sum_sub'],
    crashes_2023['NUMBER OF PERSONS KILLED']
)

# Optional: sanity check for mismatches between sub-sums and 'NUMBER OF PERSONS ‚Ä¶'
mismatch_inj = int((crashes_2023['total_injured'] != crashes_2023['NUMBER OF PERSONS INJURED']).sum())
mismatch_kil = int((crashes_2023['total_killed']  != crashes_2023['NUMBER OF PERSONS KILLED']).sum())
if mismatch_inj or mismatch_kil:
    print(f"‚ö†Ô∏è Totals mismatch in {mismatch_inj} (injured) and {mismatch_kil} (killed) rows; using sub-component sums to avoid double counting.")

# Binary outcomes
crashes_2023['fatal'] = (crashes_2023['total_killed']  > 0).astype(int)
crashes_2023['any_injury'] = (crashes_2023['total_injured'] > 0).astype(int)

# ---------- 1) VEHICLES: 2023 subset & motorcycle tagging ----------
vehicles = df2.copy()
vehicles.columns = vehicles.columns.str.strip()
if 'COLLISION_ID' not in vehicles.columns:
    raise ValueError("Vehicles table missing 'COLLISION_ID'.")

cid_2023 = set(crashes_2023['COLLISION_ID'].dropna().astype(int))
vehicles_2023 = vehicles[vehicles['COLLISION_ID'].isin(cid_2023)].copy()

veh_type_cols = [c for c in vehicles_2023.columns if ('vehicle' in c.lower() and 'type' in c.lower())]
if not veh_type_cols and 'VEHICLE_TYPE' in vehicles_2023.columns:
    veh_type_cols = ['VEHICLE_TYPE']

moto_keywords = [
    'motorcycle', 'motor bike', 'motor-bike', 'motorscooter', 'scooter',
    'moped', 'dirt bike', 'vespa', 'mini bike'
]
moto_pattern = re.compile('|'.join(map(re.escape, moto_keywords)), flags=re.I)

def is_motorcycle_row(row):
    for col in veh_type_cols:
        v = str(row.get(col, '')).lower()
        if moto_pattern.search(v):
            return True
    return False

vehicles_2023['is_motorcycle'] = vehicles_2023.apply(is_motorcycle_row, axis=1)

# Per-crash aggregation
veh_agg = vehicles_2023.groupby('COLLISION_ID').agg(
    n_vehicles=('COLLISION_ID','size'),
    any_moto=('is_motorcycle','any')
).reset_index()

# ---------- 2) Merge & keep motorcycle-involved crashes ----------
cr = crashes_2023.merge(veh_agg, on='COLLISION_ID', how='left')
cr = cr.dropna(subset=['n_vehicles'])  # remove crashes with no vehicles record
cr['n_vehicles'] = cr['n_vehicles'].astype(int)
cr['any_moto'] = cr['any_moto'].fillna(False)

cr_moto = cr[cr['any_moto']].copy()
if cr_moto.empty:
    raise ValueError("No motorcycle-involved crashes in 2023 (check vehicle type columns/keywords).")

cr_moto['self_crash'] = (cr_moto['n_vehicles'] == 1).astype(int)

# ---------- 3) 2x2 verbal test results: (self vs multi) x (fatal vs nonfatal) ----------
a = int(cr_moto.loc[cr_moto['self_crash']==1, 'fatal'].sum())
b = int((cr_moto['self_crash']==1).sum() - a)
c = int(cr_moto.loc[cr_moto['self_crash']==0, 'fatal'].sum())
d = int((cr_moto['self_crash']==0).sum() - c)
cont_fatal = np.array([[a, b],
                       [c, d]], dtype=int)

chi2_f, p_chi_f, dof_f, _ = chi2_contingency(cont_fatal, correction=False)
or_table_f = Table2x2(cont_fatal)
or_f = or_table_f.oddsratio
ci_f = or_table_f.oddsratio_confint()
p_fisher_f = fisher_exact(cont_fatal, alternative='two-sided')[1]

print("\n[Self vs Multi] √ó [Fatal vs Nonfatal]")
print(cont_fatal)
print(f"Chi-square={chi2_f:.3f}, df={dof_f}, p={p_chi_f:.4f} | Fisher p={p_fisher_f:.4f}")
print(f"Odds ratio (Self vs Multi) = {or_f:.3f}  95% CI [{ci_f[0]:.3f}, {ci_f[1]:.3f}]")

# ---------- 4) 2x2 verbal test results: (self vs multi) x (any-injury vs none) ----------
a2 = int(cr_moto.loc[cr_moto['self_crash']==1, 'any_injury'].sum())
b2 = int((cr_moto['self_crash']==1).sum() - a2)
c2 = int(cr_moto.loc[cr_moto['self_crash']==0, 'any_injury'].sum())
d2 = int((cr_moto['self_crash']==0).sum() - c2)
cont_inj = np.array([[a2, b2],
                     [c2, d2]], dtype=int)

chi2_i, p_chi_i, dof_i, _ = chi2_contingency(cont_inj, correction=False)
or_table_i = Table2x2(cont_inj)
or_i = or_table_i.oddsratio
ci_i = or_table_i.oddsratio_confint()
p_fisher_i = fisher_exact(cont_inj, alternative='two-sided')[1]

print("\n[Self vs Multi] √ó [Any-injury vs None]")
print(cont_inj)
print(f"Chi-square={chi2_i:.3f}, df={dof_i}, p={p_chi_i:.4f} | Fisher p={p_fisher_i:.4f}")
print(f"Odds ratio (Self vs Multi) = {or_i:.3f}  95% CI [{ci_i[0]:.3f}, {ci_i[1]:.3f}]")

# ---------- 5) Injury burden per crash: Poisson GLM (robust) + Mann‚ÄìWhitney ----------
glm_df = cr_moto[['self_crash','total_injured']].copy()
glm_df['intercept'] = 1.0
poisson_model = sm.GLM(glm_df['total_injured'], glm_df[['intercept','self_crash']],
                       family=sm.families.Poisson())
res = poisson_model.fit(cov_type='HC0')  # robust (sandwich) SE

coef = res.params['self_crash']
se   = res.bse['self_crash']
IRR  = np.exp(coef)
ci_lo, ci_hi = np.exp(coef - 1.96*se), np.exp(coef + 1.96*se)
p_glm = res.pvalues['self_crash']

print("\nPoisson GLM: injuries_per_crash ~ self_crash (robust SE)")
print(f"IRR (Self vs Multi) = {IRR:.3f}  95% CI [{ci_lo:.3f}, {ci_hi:.3f}],  p={p_glm:.4f}")

# Mann‚ÄìWhitney (distribution-free)
inj_self  = cr_moto.loc[cr_moto['self_crash']==1, 'total_injured']
inj_multi = cr_moto.loc[cr_moto['self_crash']==0, 'total_injured']
u_stat, p_mw = mannwhitneyu(inj_self, inj_multi, alternative='two-sided')
print(f"Mann‚ÄìWhitney U for injuries per crash: U={u_stat:.0f}, p={p_mw:.4f}")

# ---------- 6) Plots ----------
# A) Stacked bars: Fatal vs Nonfatal
grp_f = cr_moto.groupby('self_crash')['fatal'].value_counts().unstack(fill_value=0)
grp_f.index = grp_f.index.map({0:'Multi-vehicle', 1:'Self'})

plt.figure(figsize=(7,5))
plt.bar(grp_f.index, grp_f.get(1, 0), label='Fatal', color='#d62728')
plt.bar(grp_f.index, grp_f.sum(axis=1) - grp_f.get(1,0), bottom=grp_f.get(1,0), label='Nonfatal', color='#2ca02c')
plt.title('NYC 2023 | Motorcycle-involved crashes\nFatal vs Nonfatal by Self/Multi')
plt.ylabel('Number of crashes')
plt.legend()
plt.tight_layout()
plt.show()

# B) Any-injury rate with Wilson 95% CI
def wilson(k, n, alpha=0.05):
    lo, hi = proportion_confint(k, n, alpha=alpha, method='wilson')
    return lo, hi

k_vals = np.array([a2, c2], dtype=float)
n_vals = np.array([(cr_moto['self_crash']==1).sum(), (cr_moto['self_crash']==0).sum()], dtype=float)
p_vals = k_vals / n_vals
ci_w = np.array([wilson(int(k_vals[i]), int(n_vals[i])) for i in range(2)])
err_lo = p_vals - ci_w[:,0]
err_hi = ci_w[:,1] - p_vals

plt.figure(figsize=(7,5))
plt.errorbar(['Self','Multi-vehicle'], p_vals*100, yerr=[err_lo*100, err_hi*100],
             fmt='o', capsize=5, color='#1f77b4')
plt.title('NYC 2023 | Any-injury rate in motorcycle-involved crashes\n(Self vs Multi) with 95% CI')
plt.ylabel('Crashes with ‚â•1 injury (%)')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# C) Mean injuries per crash with 95% CI (normal approx)
m_self, m_multi = inj_self.mean(), inj_multi.mean()
se_self = inj_self.std(ddof=1) / np.sqrt(max(1, len(inj_self)))
se_multi= inj_multi.std(ddof=1)/ np.sqrt(max(1, len(inj_multi)))
ci_self = (m_self - 1.96*se_self, m_self + 1.96*se_self)
ci_multi= (m_multi- 1.96*se_multi, m_multi+ 1.96*se_multi)

plt.figure(figsize=(7,5))
means = [m_self, m_multi]
yerr = [[m_self - ci_self[0], m_multi - ci_multi[0]],
        [ci_self[1] - m_self,  ci_multi[1] - m_multi]]
plt.errorbar(['Self','Multi-vehicle'], means, yerr=yerr, fmt='o', capsize=5, color='#9467bd')
plt.title('NYC 2023 | Mean injuries per crash (motorcycle-involved)\nwith 95% CI')
plt.ylabel('Injuries per crash')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# =======================
# NYC 2023: Self vs Multi, using full injury/death columns
# =======================

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, fisher_exact, mannwhitneyu
from statsmodels.stats.contingency_tables import Table2x2
from statsmodels.stats.proportion import proportion_confint
import statsmodels.api as sm

# ---------- 0) CRASHES: clean & restrict to 2023 ----------
crashes = df1.copy()
crashes.columns = crashes.columns.str.strip()
crashes['CRASH DATE'] = pd.to_datetime(crashes['CRASH DATE'], errors='coerce')
crashes_2023 = crashes.loc[crashes['CRASH DATE'].dt.year == 2023].copy()

# Ensure all injury/death columns exist & numeric
col_inj_kill_all = [
    'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
    'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
    'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
    'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
]
for c in col_inj_kill_all:
    if c not in crashes_2023.columns:
        crashes_2023[c] = 0
    crashes_2023[c] = pd.to_numeric(crashes_2023[c], errors='coerce').fillna(0)

# Prefer totals from sub-components (to avoid double counting)
crashes_2023['inj_sum_sub'] = (
    crashes_2023['NUMBER OF PEDESTRIANS INJURED'] +
    crashes_2023['NUMBER OF CYCLIST INJURED'] +
    crashes_2023['NUMBER OF MOTORIST INJURED']
)
crashes_2023['kill_sum_sub'] = (
    crashes_2023['NUMBER OF PEDESTRIANS KILLED'] +
    crashes_2023['NUMBER OF CYCLIST KILLED'] +
    crashes_2023['NUMBER OF MOTORIST KILLED']
)

# If sub-components exist, use them; otherwise fall back to persons totals
crashes_2023['total_injured'] = np.where(
    crashes_2023[['NUMBER OF PEDESTRIANS INJURED','NUMBER OF CYCLIST INJURED','NUMBER OF MOTORIST INJURED']].sum(axis=1).notna(),
    crashes_2023['inj_sum_sub'],
    crashes_2023['NUMBER OF PERSONS INJURED']
)
crashes_2023['total_killed'] = np.where(
    crashes_2023[['NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST KILLED']].sum(axis=1).notna(),
    crashes_2023['kill_sum_sub'],
    crashes_2023['NUMBER OF PERSONS KILLED']
)

# Optional: sanity check for mismatches between sub-sums and 'NUMBER OF PERSONS ‚Ä¶'
mismatch_inj = int((crashes_2023['total_injured'] != crashes_2023['NUMBER OF PERSONS INJURED']).sum())
mismatch_kil = int((crashes_2023['total_killed']  != crashes_2023['NUMBER OF PERSONS KILLED']).sum())
if mismatch_inj or mismatch_kil:
    print(f"‚ö†Ô∏è Totals mismatch in {mismatch_inj} (injured) and {mismatch_kil} (killed) rows; using sub-component sums to avoid double counting.")

# Binary outcomes
crashes_2023['fatal'] = (crashes_2023['total_killed']  > 0).astype(int)
crashes_2023['any_injury'] = (crashes_2023['total_injured'] > 0).astype(int)

# ---------- 1) VEHICLES: 2023 subset & motorcycle tagging ----------
vehicles = df2.copy()
vehicles.columns = vehicles.columns.str.strip()
if 'COLLISION_ID' not in vehicles.columns:
    raise ValueError("Vehicles table missing 'COLLISION_ID'.")

cid_2023 = set(crashes_2023['COLLISION_ID'].dropna().astype(int))
vehicles_2023 = vehicles[vehicles['COLLISION_ID'].isin(cid_2023)].copy()

veh_type_cols = [c for c in vehicles_2023.columns if ('vehicle' in c.lower() and 'type' in c.lower())]
if not veh_type_cols and 'VEHICLE_TYPE' in vehicles_2023.columns:
    veh_type_cols = ['VEHICLE_TYPE']

moto_keywords = [
    'motorcycle', 'motor bike', 'motor-bike', 'motorscooter', 'scooter',
    'moped', 'dirt bike', 'vespa', 'mini bike'
]
moto_pattern = re.compile('|'.join(map(re.escape, moto_keywords)), flags=re.I)

def is_motorcycle_row(row):
    for col in veh_type_cols:
        v = str(row.get(col, '')).lower()
        if moto_pattern.search(v):
            return True
    return False

vehicles_2023['is_motorcycle'] = vehicles_2023.apply(is_motorcycle_row, axis=1)

# Per-crash aggregation
veh_agg = vehicles_2023.groupby('COLLISION_ID').agg(
    n_vehicles=('COLLISION_ID','size'),
    any_moto=('is_motorcycle','any')
).reset_index()

# ---------- 2) Merge & keep motorcycle-involved crashes ----------
cr = crashes_2023.merge(veh_agg, on='COLLISION_ID', how='left')
cr = cr.dropna(subset=['n_vehicles'])  # remove crashes with no vehicles record
cr['n_vehicles'] = cr['n_vehicles'].astype(int)
cr['any_moto'] = cr['any_moto'].fillna(False)

cr_moto = cr[cr['any_moto']].copy()
if cr_moto.empty:
    raise ValueError("No motorcycle-involved crashes in 2023 (check vehicle type columns/keywords).")

cr_moto['self_crash'] = (cr_moto['n_vehicles'] == 1).astype(int)

# ---------- 3) 2x2 verbal test results: (self vs multi) x (fatal vs nonfatal) ----------
a = int(cr_moto.loc[cr_moto['self_crash']==1, 'fatal'].sum())
b = int((cr_moto['self_crash']==1).sum() - a)
c = int(cr_moto.loc[cr_moto['self_crash']==0, 'fatal'].sum())
d = int((cr_moto['self_crash']==0).sum() - c)
cont_fatal = np.array([[a, b],
                       [c, d]], dtype=int)

chi2_f, p_chi_f, dof_f, _ = chi2_contingency(cont_fatal, correction=False)
or_table_f = Table2x2(cont_fatal)
or_f = or_table_f.oddsratio
ci_f = or_table_f.oddsratio_confint()
p_fisher_f = fisher_exact(cont_fatal, alternative='two-sided')[1]

print("\n[Self vs Multi] √ó [Fatal vs Nonfatal]")
print(cont_fatal)
print(f"Chi-square={chi2_f:.3f}, df={dof_f}, p={p_chi_f:.4f} | Fisher p={p_fisher_f:.4f}")
print(f"Odds ratio (Self vs Multi) = {or_f:.3f}  95% CI [{ci_f[0]:.3f}, {ci_f[1]:.3f}]")

# ---------- 4) 2x2 verbal test results: (self vs multi) x (any-injury vs none) ----------
a2 = int(cr_moto.loc[cr_moto['self_crash']==1, 'any_injury'].sum())
b2 = int((cr_moto['self_crash']==1).sum() - a2)
c2 = int(cr_moto.loc[cr_moto['self_crash']==0, 'any_injury'].sum())
d2 = int((cr_moto['self_crash']==0).sum() - c2)
cont_inj = np.array([[a2, b2],
                     [c2, d2]], dtype=int)

chi2_i, p_chi_i, dof_i, _ = chi2_contingency(cont_inj, correction=False)
or_table_i = Table2x2(cont_inj)
or_i = or_table_i.oddsratio
ci_i = or_table_i.oddsratio_confint()
p_fisher_i = fisher_exact(cont_inj, alternative='two-sided')[1]

print("\n[Self vs Multi] √ó [Any-injury vs None]")
print(cont_inj)
print(f"Chi-square={chi2_i:.3f}, df={dof_i}, p={p_chi_i:.4f} | Fisher p={p_fisher_i:.4f}")
print(f"Odds ratio (Self vs Multi) = {or_i:.3f}  95% CI [{ci_i[0]:.3f}, {ci_i[1]:.3f}]")

# ---------- 5) Injury burden per crash: Poisson GLM (robust) + Mann‚ÄìWhitney ----------
glm_df = cr_moto[['self_crash','total_injured']].copy()
glm_df['intercept'] = 1.0
poisson_model = sm.GLM(glm_df['total_injured'], glm_df[['intercept','self_crash']],
                       family=sm.families.Poisson())
res = poisson_model.fit(cov_type='HC0')  # robust (sandwich) SE

coef = res.params['self_crash']
se   = res.bse['self_crash']
IRR  = np.exp(coef)
ci_lo, ci_hi = np.exp(coef - 1.96*se), np.exp(coef + 1.96*se)
p_glm = res.pvalues['self_crash']

print("\nPoisson GLM: injuries_per_crash ~ self_crash (robust SE)")
print(f"IRR (Self vs Multi) = {IRR:.3f}  95% CI [{ci_lo:.3f}, {ci_hi:.3f}],  p={p_glm:.4f}")

# Mann‚ÄìWhitney (distribution-free)
inj_self  = cr_moto.loc[cr_moto['self_crash']==1, 'total_injured']
inj_multi = cr_moto.loc[cr_moto['self_crash']==0, 'total_injured']
u_stat, p_mw = mannwhitneyu(inj_self, inj_multi, alternative='two-sided')
print(f"Mann‚ÄìWhitney U for injuries per crash: U={u_stat:.0f}, p={p_mw:.4f}")

# ---------- 6) Plots ----------
# A) Stacked bars: Fatal vs Nonfatal
grp_f = cr_moto.groupby('self_crash')['fatal'].value_counts().unstack(fill_value=0)

# Normalize the counts to get proportions
grp_f_prop = grp_f.div(grp_f.sum(axis=1), axis=0)

# Rename the index for better readability
grp_f_prop.index = grp_f_prop.index.map({0: 'Multi-vehicle', 1: 'Self'})

# Create the plot
plt.figure(figsize=(7, 5))
plt.bar(grp_f_prop.index, grp_f_prop.get(1, 0), label='Fatal', color='#d62728')
plt.bar(grp_f_prop.index, grp_f_prop.get(0, 0), bottom=grp_f_prop.get(1, 0), label='Nonfatal', color='#2ca02c')

# Add labels and title
plt.title('NYC 2023 | Motorcycle-involved crashes\nProportions: Fatal vs Nonfatal by Self/Multi')
plt.ylabel('Proportion of crashes')
plt.legend()
plt.tight_layout()
plt.show()

# B) Any-injury rate with Wilson 95% CI
def wilson(k, n, alpha=0.05):
    lo, hi = proportion_confint(k, n, alpha=alpha, method='wilson')
    return lo, hi

k_vals = np.array([a2, c2], dtype=float)
n_vals = np.array([(cr_moto['self_crash']==1).sum(), (cr_moto['self_crash']==0).sum()], dtype=float)
p_vals = k_vals / n_vals
ci_w = np.array([wilson(int(k_vals[i]), int(n_vals[i])) for i in range(2)])
err_lo = p_vals - ci_w[:,0]
err_hi = ci_w[:,1] - p_vals

plt.figure(figsize=(7,5))
plt.errorbar(['Self','Multi-vehicle'], p_vals*100, yerr=[err_lo*100, err_hi*100],
             fmt='o', capsize=5, color='#1f77b4')
plt.title('NYC 2023 | Any-injury rate in motorcycle-involved crashes\n(Self vs Multi) with 95% CI')
plt.ylabel('Crashes with ‚â•1 injury (%)')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# C) Mean injuries per crash with 95% CI (normal approx)
m_self, m_multi = inj_self.mean(), inj_multi.mean()
se_self = inj_self.std(ddof=1) / np.sqrt(max(1, len(inj_self)))
se_multi= inj_multi.std(ddof=1)/ np.sqrt(max(1, len(inj_multi)))
ci_self = (m_self - 1.96*se_self, m_self + 1.96*se_self)
ci_multi= (m_multi- 1.96*se_multi, m_multi+ 1.96*se_multi)

plt.figure(figsize=(7,5))
means = [m_self, m_multi]
yerr = [[m_self - ci_self[0], m_multi - ci_multi[0]],
        [ci_self[1] - m_self,  ci_multi[1] - m_multi]]
plt.errorbar(['Self','Multi-vehicle'], means, yerr=yerr, fmt='o', capsize=5, color='#9467bd')
plt.title('NYC 2023 | Mean injuries per crash (motorcycle-involved)\nwith 95% CI')
plt.ylabel('Injuries per crash')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
